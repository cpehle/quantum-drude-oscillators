```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy
import math
```

```python
class Walker(object):
    def __init__(self, numid, pos):
        self.pos = pos
        self.numid = numid
        
    def V(self):
        return 0.5 * self.pos.dot(self.pos)
    
    def step(self, dt, target_energy):
        self.pos += np.sqrt(dt) * scipy.random.randn(3)
        q = math.exp(-dt * (self.V() - target_energy))
        ncopies = int(q + scipy.random.random())
        #print "%.5f %.5f %.5f" % (self.V(), target_energy, q), ncopies, self.numid
        return ncopies
    
class DMC(object):
    def __init__(self, target_nwalkers):
        self.walkers = []
        for ii in xrange(target_nwalkers):
            self.walkers.append(Walker(ii, 
                                       np.array((np.random.rand(3)-0.5), dtype=np.float64)))
        self.target_energy = np.float64(0.0)
        self.target_nwalkers = np.float64(target_nwalkers)
        
    def timestep(self, dt):
        new_walkers = []
        for ii,walker in enumerate(self.walkers):
            ncopies = walker.step(dt, self.target_energy)
            for jj in xrange(ncopies):
                new_walkers.append(Walker(walker.numid, walker.pos))
        #print
        self.walkers = new_walkers
        if not self.walkers:
            raise
        self.target_energy += (math.log(self.target_nwalkers) - math.log(len(self.walkers)))/10.
        return self.target_energy
```

```python
dd = DMC(1600)
es = []
for ii in xrange(1900):
     es.append(dd.timestep(0.005))
```

```python
plt.plot(es)
```
